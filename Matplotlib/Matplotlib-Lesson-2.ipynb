````python
import pandas as pd
````

````python
sums = list(tips.groupby('day')['total_bill'].sum())
sums
````

````python
day = tips.groupby("day").sum().index   # tips.day.unique()
day
````

````python
fig, ax = plt.subplots()
ax.bar(day, sums)
plt.show()
````

````python
fig, ax = plt.subplots()
ax.barh(day, sums)
plt.show()
````

````python
tipp = list(tips.groupby('day')['tip'].sum())
tipp
````

````python
fig, ax = plt.subplots()
ax.bar(day, tipp)
plt.show()
````

````python
# Optional hard to coding
fig, ax = plt.subplots(figsize=(10,5))
p = np.arange(len(day))
width = 0.2

ax.bar(p - width/2, sums, width=0.2, label="total_bill", color='g')
ax.bar(p + width/2, tipp, width=0.2, label="tip", color='r')

ax.set_xticks(p)
ax.set_xticklabels(day)
plt.legend()
plt.show()
````

````python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
````

````python
plt.pie(y, labels=mylabels, radius=2, labeldistance = 1, autopct='%0.2f%%')
plt.show()
````

````python
plt.pie(y, labels=mylabels, radius=2, labeldistance = 1, autopct='%0.2f%%', 
        startangle=90, explode=[0.2,0,0,0])
plt.show()
````

````python
plt.figure(figsize=(10,8))
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']

explode_list = [0, 0, 0.1, 0.1] # ratio for each continent with which to offset each wedge.
y = np.array([25, 35, 25, 15])

mylabels = ["Bananas", "Apples", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, labeldistance=0.7, autopct="%.1f", startangle=90, 
        shadow=True, pctdistance=1.1, colors=colors_list, explode=explode_list)
plt.legend()
plt.show()
````
