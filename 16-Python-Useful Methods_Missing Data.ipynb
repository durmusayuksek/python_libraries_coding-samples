### map(), apply() ve applymap() methods

````python
import pandas as pd
````

````python
df = pd.read_csv('http://bit.ly/kaggletrain')
````

````python
df.head()
````

```
apply() : Apply a function along an axis of the DataFrame. Series in given axis are passed to the function.
Row or Column Wise Function Application: apply()
Element wise Function Application: applymap()
Function application on Series : map()
```

````python
df['sex_num'] = df.Sex.map({'male' : 1, 'female': 0})
````

````python
df.loc[0:4, ['Sex', 'sex_num']]
````

````python
df['name_length'] = df.Name.apply(lambda x : len(x))
                          # .apply(len) de diyebilirdim
````

````python
df.loc[:4, ['Name', 'name_length']]
````

````python
import numpy as np
````

````python
df['fare_rounded'] = df.Fare.apply(np.ceil)
````

````python
df.loc[:4, ['Fare', 'fare_rounded']]
````

````python
import re
df['last_name'] = df.Name.apply(lambda x : re.findall('(\S+),', x))
````

````python
df.loc[:4, ['Name', 'last_name']]
````

````python
df['last_name2'] = df.Name.str.split(',').apply(lambda x : x[0]).head()
````

````python
df.loc[:4, ['Name', 'last_name2']]
````

````python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
````

````python
drinks.head()
````

````python
drinks.loc[:, 'beer_servings':'wine_servings'].apply(max, axis = 0)
````

````python
drinks.loc[:, 'beer_servings':'wine_servings'].apply(max, axis = 1)
````

````python
drinks.loc[:, 'beer_servings':'wine_servings'].apply(np.argmax, axis = 1)
# apply() hem bir series metodu olabilir, hem de bir dataframe metodu olabilir
# eger dataframe metodu olarak kullanilacaksa, axis belirtmemiz gerekiyor
````

````python
drinks.loc[:, 'beer_servings':'wine_servings'].applymap(float)
# applymap() da bir dataframe metodudur. applymap()'de axis belirtmeye gerek yoktur, cunku butun elemanlara uygular
````

````python
drinks.loc[:, 'beer_servings':'wine_servings'] = drinks.loc[:, 'beer_servings':'wine_servings'].applymap(float)
# applymap() ile bu sekilde datada degisiklik de yapabilirim
````

````python
drinks
````

### Missing Data

````python
import pandas as pd
````

````python
df = pd.read_csv('Bengaluru.csv')
````

````python
df.head()
````

````python
df.shape
````

````python
df.isnull() # bir cell'in null olup olmadigini gosteriyor. null ise True veriyor
````

````python
df.isnull().sum()
````

````python
df.isnull().sum().sum()
````

#### 1. Filling Null Values

````python
df2 = df.fillna(value = 0)
df2
````

````python
df2.isnull().sum()
````

#### 2. Filling Null Values with the Previous Valid Value

````python
df3 = df.fillna(method = 'pad')
df3 # shift tab yapinca method kismi ile ilgili aciklama cikiyor. pad bir onceki valid degerle ayni degeri atiyor
````

#### 3. Filling Null Values with the Next Valid Value

````python
df4 = df.fillna(method = 'bfill')
df4
````

````python
df5 = df.fillna(method = 'pad', axis = 1)
df5
````

#### 4. Filling Different Null Values in Different Columns

````python
df6 = df.fillna({'society':'abcd', 'balcony':'defg'})
df6
````

#### 5. Filling Null Values with the Mean

````python
df7 = df.fillna(value = df['balcony'].mean())
df7
````

#### 6. Dropping Null Values

````python
df8 = df.dropna() # null bulunan tum satirlari siliyor
df8
````

````python
df9 = df.dropna(how = 'all')
df9
````

#### 7. Replacing Null Values

````python
import numpy as np
df10 = df.replace(to_replace = np.nan, value = 1001)
df10
````

````python
df11 = df.replace(to_replace = 3, value = 1000) # replace'in fillna'dan farki istedigimiz her degeri degistirebiliyoruz.
df11
````

#### 8. Interpolate

````python
df['balcony'] = df['balcony'].interpolate(method = 'linear')
df
````
