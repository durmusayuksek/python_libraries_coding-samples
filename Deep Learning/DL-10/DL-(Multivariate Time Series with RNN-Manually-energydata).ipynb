# Dataset Info

Appliance Energy Usage: A Multivariate Time Series Forecasting Example

Experimental data used to create regression models of appliances energy use in a low energy building.
Data Set Information:

The data set is at 10 min for about 4.5 months. The house temperature and humidity conditions were monitored with a ZigBee wireless sensor network. Each wireless node transmitted the temperature and humidity conditions around 3.3 min. Then, the wireless data was averaged for 10 minutes periods. The energy data was logged every 10 minutes with m-bus energy meters. Weather from the nearest airport weather station (Chievres Airport, Belgium) was downloaded from a public data set from Reliable Prognosis (rp5.ru), and merged together with the experimental data sets using the date and time column. Two random variables have been included in the data set for testing the regression models and to filter out non predictive attributes (parameters).
Original source of the dataset:

http://archive.ics.uci.edu/ml/datasets/Appliances+energy+prediction

### Scale Data

```python
df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler

f_columns = df.columns[1:]

f_transformer = RobustScaler() # appliances disindaki diger feature'lar icin kullaniliyor
target_transformer = RobustScaler() # sadece appliances icin kullaniliyor

train.loc[:, f_columns] = f_transformer.fit_transform(train[f_columns])
train['Appliances'] = target_transformer.fit_transform(train[['Appliances']])

test.loc[:, f_columns] = f_transformer.transform(test[f_columns])
test['Appliances'] = target_transformer.transform(test[['Appliances']])

# ben appliances'i predict yapmak istedigim icin burada appliances'i diger feature'lardan
# ayirarak fit ve transform islemini uyguluyoruz
# bunun sebebi islem sonucunda inverse transform islemini appliances uzerinde uygulayacagimiz
# icin bu asamada sorun cikmamasi icin appliances'i burada digerlerinden ayiriyoruz
```

### Define a Function for Creating Time Steps 

```python
def create_dataset(X, y, time_steps=1):
    """ Create data sequence
    
    Arguments:
        * X: time-series data
        * y: Count "cnt" value
        * time_steps: Used to create input sequence of timesteps
    
    Returns:
        * input_sequence: Numpy array of sequences of time-series data
        * output: Numpy array of output i.e. next value for respective sequence
    
    """
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)        
        ys.append(y.iloc[i + time_steps])
    return np.array(Xs), np.array(ys)
```

```python
time_steps = 24 # length (cok uzun olmasi tahminin gucunu zayiflatir)

# reshape to [samples, time_steps, n_features]

X_train, y_train = create_dataset(train, train.Appliances, time_steps)
X_test, y_test = create_dataset(test, test.Appliances, time_steps)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

# 3098 tane 24 satira 10 sutunluk array'im var

# 3098 elemandan olusan tek bir sutun
```

## Model Building with LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
n_features = X_train.shape[2]
```

```python
model = Sequential()

model.add(Bidirectional(LSTM(units=100, input_shape=(time_steps, n_features))))
# model 24'e 10'luk veri bekleyecek
# bidirectional'da her bir noron datayi 1'den 24'e kadar alacak, sonra 24'ten 1'e kadar alacak,
# sonrasinda her ikisi sonucunda elde edilen agirlik degerlerinin ortalamasi alinacak
#model.add(Dropout(rate=0.2))

model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mse')
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', patience = 5, restore_best_weights=True)

# restore_best_weights ile en iyi skorlarin elde edildigi adimin degerlerini alir
```

```python
model.fit(X_train, y_train, 
          epochs=30, 
          batch_size=1, 
          validation_split=0.3,
          # shuffle=False,
          callbacks=[early_stop])

# shuffle=True her epoch'ta datami kariyor ve modele sokuyor

# burada yukarida olusturdugumuz 24'lu paketlerden bahsediyoruz. Onlari shuffle=True dersek
# paketleri kariyor, paketlerin icini karmiyor.
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot();
```

## Evaluate on Test Data

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\nmae:", mae, "\nmse:",mse, "\nrmse:",rmse)
```

```python
y_pred = model.predict(X_test)
```

```python
eval_metrics(y_test, y_pred)
```

### reshape to [samples, time_steps, n_features]

```python
X, y = create_dataset(full, full.Appliances, time_steps)

print(X.shape, y.shape)
```

### Create Final model with full data

```python
model = Sequential()

model.add(Bidirectional(LSTM(units=100, input_shape=(time_steps, n_features))))

#model.add(Dropout(rate=0.2))

model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X, y, 
          epochs=15, 
          batch_size=1, 
         # shuffle=False
         )
```

### First way : after each prediction, reel value will be use for next prediction¶

```python
forecast = []
# Replace periods with whatever forecast length you want
periods = 24

first_eval_batch = full.values[-time_steps:]
current_batch = first_eval_batch.reshape((1, time_steps, full.shape[1]))
features = df_fake.values.reshape((df_fake.shape[0], 1, df_fake.shape[1]))

for i in range(periods):
    
    # get prediction 1 time stamp ahead 
    current_pred = model.predict(current_batch)[0][0]
    
    # store prediction
    forecast.append(current_pred) 
    new = features[i][0].tolist()
    new.insert(0,reel[i][0]) # burada yine gercek degerle bir sonraki tahmini yapiyor
    new = np.array(new).reshape(1,1,full.shape[1])
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], new, axis = 1)
```
