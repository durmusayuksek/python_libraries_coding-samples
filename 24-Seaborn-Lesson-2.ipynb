```python
tips.describe()
```

### Countplot

```python
tips.day.value_counts(sort=False)
```

```python
sns.countplot(x='day', data=tips);
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x='day', data=tips)
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+0.8));
    # matplotlib.pyplot.annotate(text, xy, *args, **kwargs)
```

```python
for p in ax.patches:
    print(p)
```

```python
sns.countplot(x='day', data=tips, hue='sex');
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x='day', data=tips, hue='sex')
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.125, p.get_height()+0.5));
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x='day', data=tips, hue="sex")
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+1))
    ax.text(p.get_x()+0.1, p.get_height()-3,str(round(p.get_height()/tips.day.count(),2)) )    
    # yüzdelik ekledik
    
# burada annotate ile ilk olarak neyi yazdirmak istiyorsak onu istiyoruz, sonrasinda
# yer belirliyoruz. Ilk olarak x'i sonrasinda y'yi belirtiyoruz
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x='day', data=tips, hue="sex")
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+0.7))
    ax.annotate(round(p.get_height()/tips.day.count(),2), (p.get_x()+0.1,
                                   p.get_height()-4))
```
