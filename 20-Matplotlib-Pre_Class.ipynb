```
Matplotlib is an amazing visualization library for the Python programming language and its numerical mathematics extension NumPy. Matplotlib was originally written and introduced by John Hunter in the year 2002.  He created it to use MatLab's plotting capabilities in Python. To use Matplotlib plots on python we use matplotlib.pyplot module. It is a collection of command style functions that make Matplotlib work like MatLab.
```

### Matplotlib - Subplots

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

#### Functional

````python
x = np.linspace(0,10,10)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = np.sqrt(x)
````

````python
plt.figure(tight_layout=True)
plt.subplot(1,4,1)
plt.plot(x,y1,'ro')
plt.title('$y_1=x$')

plt.subplot(1,4,2)
plt.plot(x,y2,'g--')
plt.title('$y_2=x^2$')

plt.subplot(1,4,3)
plt.plot(x,y3,'b.')
plt.title('$y_3=x^3$')

plt.subplot(1,4,4)
plt.plot(x,y4,'-')
plt.title('$y_4=sqrt(x)$')
````

````python
plt.figure(tight_layout=True)
plt.subplot(2,2,1)
plt.plot(x,y1,'ro')
plt.title('$y_1=x$')

plt.subplot(2,2,2)
plt.plot(x,y2,'g--')
plt.title('$y_2=x^2$')

plt.subplot(2,2,3)
plt.plot(x,y3,'b.')
plt.title('$y_3=x^3$')

plt.subplot(2,2,4)
plt.plot(x,y4,'-')
plt.title('$y_4=sqrt(x)$')
````

### Matplotlib Plotting Type

#### Bar Chart

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
````

````python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90,136,89,27]
profit = [40,2,34,12]
````

````python
xpos = np.arange(len(company))
xpos
````

````python
plt.xticks(ypos, company)
plt.ylabel('revenue')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2, revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2, profit, width=0.4, label='Profit')
plt.legend()
plt.show()
````

````python
plt.yticks(ypos, company)
plt.ylabel('revenue')
plt.title('US Tech Stocks')
plt.barh(xpos-0.2, revenue, height=0.4, label='Revenue')
plt.barh(xpos+0.2, profit, height=0.4, label='Profit')
plt.legend()
plt.show()
````

#### Histogram

````python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
````

````python
plt.hist(blood_sugar, bins=3, rwidth=0.95)
````

````python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
````

````python
plt.hist(blood_sugar, bins=4, rwidth=0.95, color='g', histtype='step')
````

````python
blood_sugar_man = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_woman = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('Sugar Range')
plt.ylabel('Total Number of Patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_man, blood_sugar_woman], bins=3, 
         rwidth=0.90, color=['g', 'r'], label=['men','women'])
plt.legend()
plt.show()
````

````python
plt.xlabel('Sugar Range')
plt.ylabel('Total Number of Patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_man, blood_sugar_woman], bins=3, 
         rwidth=0.90, color=['g', 'r'], label=['men','women'], orientation='horizontal')
plt.legend()
plt.show()
````

#### Pie Chart

````python
exp_vals = [1400,600,300,410,250]
exp_labels = ['Home Rent', 'Food', 'Phone/Internet Bill', 'Car', 'Other Utilities']
````

````python
plt.pie(exp_vals, labels=exp_labels)
plt.show()
````

````python
plt.pie(exp_vals, labels=exp_labels, radius=2)
plt.show()
````

````python
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%', shadow=True) 
# eger percentage decimal olsun dersem 0.1, 0.2, eger percentage tam olsun diyorsam 0.0
plt.show()                                                       
````

````python
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%', explode=[0,0,0,0,0]) 
plt.show() 
````

````python
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%', explode=[0.1,0,0,0,0]) 
plt.show() 
````

````python
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%', explode=[0.1,0,0.5,0,0.3]) 
plt.show()
````

````python
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%', startangle=90) 
plt.show() # Ilk degisken olan home rent 90 derecelik aciyla bagliyor ve counter clockwise ilerliyor digerleri
````
