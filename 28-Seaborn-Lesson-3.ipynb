```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.catplot(x='pclass', y='survived', data=titanic, kind='bar');
```

```python
sns.catplot(x='pclass', y='survived', data=titanic, kind='bar', estimator=np.count_nonzero);
```

```python
sns.catplot(x='pclass', y='survived', data=titanic, kind='bar', estimator=np.count_nonzero,
           hue='sex');
```

```python
sns.catplot(x='pclass', y='survived', data=titanic, kind='bar', estimator=np.count_nonzero,
           hue='sex', col='alone', row='embarked');
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.catplot(x='day', y='total_bill', kind='box', data=tips);
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips);
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex');
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg');
```

```python
sns.pairplot(tips);
```

```python
g = sns.pairplot(tips, hue='sex', palette='twilight');
```

```python
g.palette
```

```python
sns.pairplot(tips, hue='sex', palette='seismic', diag_kind='hist');
```

```python
g = sns.PairGrid(tips)
g.map(sns.scatterplot);
```

```python
g = sns.PairGrid(tips)
g.map(sns.scatterplot);
```

```python
g = sns.PairGrid(tips)
g.map_upper(sns.scatterplot, color='g')
g.map_diag(sns.boxplot, color='xkcd:golden')
g.map_lower(sns.kdeplot, color='purple');
```

```python
a = sns.pairplot(tips);
```

```python
a = sns.pairplot(tips)
a.map_lower(sns.kdeplot, color='purple')
a.map_upper(sns.kdeplot, color='g');
```

```python
g = sns.FacetGrid(data = tips, col="time", row="smoker")
g = g.map(sns.histplot, "total_bill")
```

```python
g = sns.FacetGrid(data = tips, col="time", row="smoker")
g = g.map(sns.regplot, "total_bill", 'tip')
```
