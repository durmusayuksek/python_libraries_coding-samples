MINI CHALLENGE #3:
- Plot the pie chart for the same stocks assuming equal allocation
- Explode Amazon and Google slices


```python
plt.figure(figsize=(10,8))
colors_list = ['blue', 'yellowgreen', 'lightcoral']

explode_list = [0, 0, 0.1] # ratio for each continent with which to offset each wedge.
x = 1
y = np.array([x/3, x/3, x/3])

mylabels = ["NFLX", "FB", "TWTR"]

plt.pie(y, labels = mylabels, labeldistance=0.7, autopct="%.1f", startangle=90, 
        shadow=True, pctdistance=1.1, colors=colors_list, explode=explode_list)
plt.legend();
```

# 7. PLOT 3D PLOTS

MINI CHALLLENGE #7:
- Create a 3D plot with daily return values of Twitter, Facebook and Netflix


```python
fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(daily_return_df.NFLX, daily_return_df.TWTR, daily_return_df.FB, color='r', 
           marker='o', edgecolor='black', s=100)

ax.set_xlabel('NFLX', color='r', fontdict={'fontsize':20})
ax.set_ylabel('TWTR', color='r', fontdict={'fontsize':20})
ax.set_zlabel('FB', color='r', fontdict={'fontsize':20});
```

# 8. SEABRON SCATTERPLOT & COUNTPLOT

```python
# Seaborn is a visualization library that sits on top of matplotlib
# Seaborn offers enhanced features compared to matplotlib
# https://seaborn.pydata.org/examples/index.html

# import libraries 
import seaborn as sns # Statistical data visualization
```

```python
# Import Cancer data drom the Sklearn library
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

```python
# Create a dataFrame named df_cancer with input/output data
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], 
                                                                                      ['target']))
```

```python
# Check out the head of the dataframe
df_cancer.head()
```

```python
# Check out the tail of the dataframe
df_cancer.tail()
```

```python
# Plot scatter plot between mean area and mean smoothness
plt.figure(figsize=(10,6))
sns.scatterplot(x='mean area', y='mean smoothness', data=df_cancer);
```

```python
# Let's print out countplot to know how many samples belong to class #0 and #1
plt.figure(figsize=(10,6))
sns.countplot(x='target', data=df_cancer);
```

```python
df_cancer.info()
```

```python
df_cancer[['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness']]
```

MINI CHALLENGE #8:
- Plot the scatterplot between the mean radius and mean area. Comment on the plot


```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='mean radius', y='mean area', data=df_cancer);
```

```python
# There is a strong linear relationship between mean radius and mean area. 
```

# 9. SEABORN PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS


```python
sns.pairplot(df_cancer[['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness', 'target']], 
             hue='target', diag_kind='kde');
```

```python
# Strong correlation between the mean radius and mean perimeter, mean area and mean primeter
```

```python
# plot the distplot 
# Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)
# KDE is used to plot the Probability Density of a continuous variable. 

sns.distplot(df_cancer['mean radius']);
```

MINI CHALLENGE #9:
- Plot two separate distplot for each target class #0 and target class #1

```python
class_0_df = df_cancer[df_cancer["target"]==0]
class_0_df
```

```python
class_1_df = df_cancer[df_cancer["target"]==1]
class_1_df
```

```python
plt.figure(figsize=(12,7))
sns.distplot(class_0_df['mean radius'], color='b');
sns.distplot(class_1_df['mean radius'], color='r');
```
