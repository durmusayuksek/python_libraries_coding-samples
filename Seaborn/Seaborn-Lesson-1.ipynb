```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Scatter Plot

```python
figg = sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', s=200, style='sex');
```

```python
a = figg.get_figure()
```

```python
a.savefig('scatter.png')
```

```python
plt.figure(figsize=(12,8))

ax = sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
plt.figure(figsize=(12,8))

ax = sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
ax.set(xlabel="Total Bill", ylabel='Tips', title='Some Title');
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', marker='*', s=200, data=tips);
```

```python
sns.scatterplot(x='total_bill', y='tip', size='sex', s=100, data=tips);
```

```python
sns.scatterplot(x='total_bill', y='tip', size='sex', s=100, alpha=0.4, data=tips);
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='total_bill',y='tip', data=tips,style='sex',hue='sex',s=100)
# Call savefig in the same cell
plt.savefig('example_scatter.jpg')
```
