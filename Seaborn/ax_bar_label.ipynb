```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
plt.figure(figsize=(16,8))
ax = sns.countplot(data=iris, x='sepal_length')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(16,8))
ax = sns.barplot(data=iris, x='sepal_length', y='sepal_width', ci=None, hue='species')
for container in ax.containers:
    ax.bar_label(container)
```

```python
plt.figure(figsize=(16,8))
ax = sns.barplot(data=iris, x='sepal_length', y='sepal_width', ci=None, hue='species')
ax.bar_label(ax.containers[0])
plt.show()
```
