#### Violinplot

```
Draw a combination of boxplot and kernel density estimate.

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

This can be an effective and attractive way to show multiple distributions of data at once, but keep in mind that the estimation procedure is influenced by the sample size, and violins for relatively small samples might look misleadingly smooth.
```

````python
cars.head()
````

````python
sns.violinplot(cars.displacement);
````

````python
sns.violinplot(x=cars.cylinders, y=cars.displacement);
````

````python
sns.violinplot(x='cylinders', y='displacement', data=cars);
````

````python
sns.violinplot(x='cylinders', y='displacement', data=cars, hue='origin');
````

````python
sns.violinplot(x='cylinders', y='displacement', hue='origin', data=cars[cars.origin.isin(['japan', 'europe'])])
plt.legend(loc='best');
````

````python
sns.violinplot(x='cylinders', y='displacement', hue='origin', 
               data=cars[cars.origin.isin(['japan', 'europe'])], split=True)
plt.legend(loc=2);
# burada solda 4 ve 6 silindir icin japonya kde'si, sagda 4 ve 6 silindir icin avrupa kde'si ortaya cikiyor
# split kullanabilmemiz icin iki kategori olmak zorunda
````

````python
sns.violinplot(x='cylinders', y='displacement', hue='origin', 
               data=cars[cars.origin.isin(['japan', 'europe'])], split=True, inner='quartiles')
plt.legend(loc=2);
````

````python
sns.violinplot(x='cylinders', y='displacement', hue='origin', 
               data=cars[cars.origin.isin(['japan', 'europe'])], split=True, scale='count')
plt.legend(loc=2);

# normal sartlarda her iki ulke icin de kde'larin buyuklukleri ayni, ancak scale kullanilarak
# buradaki gibi her bir ulke icin araba sayilariyla orantili bir kde de uretilebiliyor
# mesela there are fewer observations for the european 6 cylinders than japanese 6 cylinders
````

````python
sns.violinplot(x='cylinders', y='displacement', hue='origin', 
               data=cars[cars.origin.isin(['japan', 'europe'])], split=True, scale='count', scale_hue=False)
plt.legend(loc=2);
# aslinda observation'larda cok daha fazla 4 cylinder car var
````

````python
cars[cars.origin.isin(['japan', 'europe'])].groupby('cylinders').origin.value_counts()
````

````python
sns.violinplot(x=cars.origin, y=cars.displacement, order=['japan', 'europe', 'usa']);
````

````python
sns.violinplot(x=cars.cylinders, y=cars.displacement, linewidth=3);
````

````python
sns.violinplot(x=cars.cylinders, y=cars.displacement, bw=0.4);
# default bw is 0.4
````
