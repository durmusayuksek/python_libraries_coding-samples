### Filtering with Groupby

```python
drinks
```

```python
drinks['continent'].value_counts()
```

```python
drinks.groupby('continent').std()
```

```python
def filtering(x):
    return x['wine_servings'].std() < 22
```

```python
drinks.groupby('continent').filter(filtering) # sadece asya ulkelerini gosteriyor, filtreledik
```

### Transform()

```python
drinks
```

```python
drinks.groupby('continent').mean()[['beer_servings', 'spirit_servings', 'wine_servings']]
```

```python
drinks.groupby('continent').transform(lambda x : x - x.mean())
# burada yapilan sey her bir row'daki degerleri alir, kendi column'unun mean'ininden cikarir ve yazdirir
```

### Apply()

```python
def normalize(x):
    x['beer_servings'] /= x['wine_servings'].sum()
    return x
```

```python
drinks.groupby('continent').apply(normalize)
```

### Creating your own splits for groupby()

```python
df1 = pd.DataFrame({'a': [1,2,3,4,44,32], 'b':[5,6,7,8,11,23], 'c': ['A', 'B', 'A', 'B', 'E', 'C']})
df1
```

```python
liste = [0,1,2,0,3,2]
df1.groupby(liste).sum()
```

```python
df2 = df1.set_index(['c'])
df2
```

```python
df2.groupby(str.lower).mean()
```

```python
data = pd.DataFrame([('bird', 389.0),
                     ('bird', 40.0), ('bird', 405.0),
                     ('mammal', 80.2),
                     ('mammal', 12.0),('mammal', 58)],
                     index=['falcon', 'parrot', 'eagle', 'lion', 'monkey', 'leopard'],
                    columns=('class', 'speed'))
data
```
